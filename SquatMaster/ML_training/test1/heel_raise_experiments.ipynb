```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('/home/rafik/PROJECTS/pose1/Material/hill_raise/cropped_feet2/right_foot/right_foot2_315.jpg')

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Calculate optimal threshold value using Otsu's thresholding
_, thresholded = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Morphological operations (closing)
kernel = np.ones((15, 15), np.uint8)
closed = cv2.morphologyEx(thresholded, cv2.MORPH_CLOSE, kernel)

# Find contours
contours, _ = cv2.findContours(closed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Iterate through contours and analyze shadows
for contour in contours:
    # Analyze shadow for each contour (you may need to adapt this part)
    # Use the information about the shadow to detect whether the heel is raised.

    # For example, check the thickness of the contour:
    contour_thickness = cv2.contourArea(contour)
    # if contour_thickness < 1:
    #     print("Heel is raised")
    # else:
    #     print("Not raised")

cv2.drawContours(image, [contour], -1, (0,255,0), 2)
cv2.imshow("Contour Overlay", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
